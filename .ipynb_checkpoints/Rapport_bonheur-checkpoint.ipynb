# enlevons les étoiles quand celles ci sont présentes dans la dfcomp

```python
def enleveetoile(df,col):
    for i in range(df.shape[0]):
        if df[col][i][-1]=='*':
           df[col][i]=df[col][i][:-1]
    return df
    
        
```

```python
dfcomp=enleveetoile(dfcomp,'2019')
dfcomp=enleveetoile(dfcomp,'2020')
dfcomp=enleveetoile(dfcomp,'2021')
dfcomp=enleveetoile(dfcomp,'2022')
```

# Changeons l'écriture de Trinidad and Tobago

```python
for col in dfcomp.columns:
    for i in range(dfcomp.shape[0]):
        if dfcomp[col][i]=='Trinidad and Tobago':
            dfcomp[col][i]='Trinidad & Tobago'
            
```

# Regardons les entrées dans le top 50 en 2020 

```python
def entreetop50(df,col1, col2):
    dftop=df.head(50)
    listecol1=[]
    listecol2=[]
    entree=[]
    for i in range(dftop.shape[0]):
        listecol1.append(dftop[col1][i])
        listecol2.append(dftop[col2][i])
    for i in range(len(listecol1)):
        if listecol1[i] not in listecol2 :
            entree.append(listecol1[i])
    return entree
        
```

```python
print('Les pays qui sont entré dans le top 50 du classement en 2020 sont : '+' '.join(entreetop50(dfcomp,'2021', '2022')))
```
