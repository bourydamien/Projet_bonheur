```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from scipy import stats
```

```python
df=pd.read_csv('bonheur_avec_meanyearschooling.csv')
```

```python
df.head()
```

# Regardons les corr√©lations 

```python
m = np.triu(df.corr())
sns.heatmap(df.corr(), annot =True, mask=m)
```

```python

sns.displot(data=df, x='Score', y='mean_year_school_2018')
plt.title("Lien entre le nombre d'ann√©es pass√©es √† l'√©cole et le score du Bonheur")
```

```python
sns.set(rc = {'figure.figsize':(10,10)})
sns.scatterplot(data=df, y='Score', x='HDI Rank', color='red', marker='^', hue='Region')
plt.title("Lien entre l'indicateur HDI et le bonheur")

```
